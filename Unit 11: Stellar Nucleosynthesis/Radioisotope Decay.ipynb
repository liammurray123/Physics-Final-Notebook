# Radiactive Decay Simulation
(This code is based on the library Radioactivedecay at https://radioactivedecay.github.io/).

Below, create an Inventory of radionuclides and decay patterns.

Use the command rd.Inventory({'Element Here': Value}, 'Unit') to provide the initial conditions.

Units can be in: Activity (Bq, Ci, dpm), Mass (g, kg), Mole, (mol, kmol), or number of nuclei (num)

Use .decay(value, 'units') to calculate the decay products. The usings can be: 'Î¼s', 'ms', 's', 'm', 'h', 'd', 'y'

Use .cumulative_decays() to calculate the total number of atoms of each radionuclide that decay over the decay time period

```python
Ni56 = rd.Inventory({'Ni-56': 100}, 'kg')
Ni56_products = Ni56.decay(10, 'd')
Ni56_products.masses('kg')
```

```python
Ni56_products.plot(200, 'd', yunits='kg')
plt.show()
```
