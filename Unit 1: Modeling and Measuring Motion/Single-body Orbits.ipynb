## Single Body Orbits

```python
# you can uncomment the next line if the ipympl library for interactive plots works on your system
%matplotlib ipympl
import matplotlib.pyplot as plt
import numpy as np

# Initialize some variables
t = 0.0             # Start the simulation at t = 0
dt = 60             # TODO: Set the time step (s) !!!!!!!! [Set to one hour]
x =  0              # TODO: Set the initial x position (m)  !!!!!!!!
y =  384000e3       # TODO: Set the initial y position (m) !!!!!!!!
v_x = -1000         # TODO: Set the initial x velocity (m/s)!!!!!!!!
v_y = 0             # TODO: Set the initial y velocity (m/s)!!!!!!!! Vmoon = 1023.056
G = 6.67e-11        # Universal gravitation constant
M = 5.97219e24      # mass of Earth in kg
ntot = 50000        # TODO: number of iterations !!!!!!!!

# initalize some lists for plotting
xposition = []
yposition = []
n = 0 # counter

# Run the simulation in a while loop
while n <= ntot:
    xposition.append(x)     # store the current x position
    yposition.append(y)     # store the current y position

    r2 = x**2 + y**2
    r = r2**0.5             # distance from Earth
    a_x = -G*M/r2 * x/r     # Update the acceleration in the x direction
    a_y = -G*M/r2 * y/r     # Update the acceleration in the y direction
    v_x = v_x + a_x*dt      # Update the velocity in the x direction
    v_y = v_y + a_y*dt      # Update the velocity in the y direction
    x = x + v_x*dt          # Update the position in the x direction
    y = y + v_y*dt          # Update the position in the y direction
    t = t + dt              # Increment the time by the time step
    n = n + 1

# Make the plots
# print(xposition)
# create circle for the Earth

rE = 6.38*np.sqrt(1.0)  # radius of Earth in Mm
theta = np.linspace(0, 2*np.pi, 100)
xE = rE*np.cos(theta)
yE = rE*np.sin(theta)

xMm = [i/1e6 for i in xposition]
yMm = [i/1e6 for i in yposition]
plt.figure()
plt.plot(xE,yE, ".", label="Earth")
plt.plot(xMm,yMm, ".", label="Moon")
plt.axis('square')
plt.title("Orbital Path of Moon")
plt.xlabel("x (1000 km)")
plt.ylabel("y (1000 km)")
plt.legend()
plt.grid(True)
plt.show()
```
