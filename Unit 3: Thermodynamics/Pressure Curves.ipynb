## Drawing Isotherms

$$
PV = Nk_BT \\ \text{where} \ T = const.
$$

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Initial Conditions #######################################################################################################

rho = 1.225 # Density, kg/m^2
Vt = 1**3 # V Volume of container, m^3
Mt = rho * Vt # Total mass of gas, Kg
Mm = 28.97 * 1.66E-27 # Atomic mass unit, in Kg
R = 8314.3 # Gas constant
Kb = 1.380649e-23 # Boltzmann's constant
N = Mt / Mm


K = 273.15 # 0 deg K

plt.figure()

for Tcel in range(-100,201,100): #temperatures in Celcius
    T = Tcel + K # Temp in Kelvin
    V = np.linspace(0.19, 1.5)
    Pisotherm = (N*Kb*T)/V
    plt.plot(V, Pisotherm/1000, label=(str(Tcel) + "C"))

# Plot Data
plt.plot(1,101.325, 'x', label=("15 C, sea level"))
plt.xlabel("Volume (m$^2$)")
plt.ylabel("Pressure (kPa)")
plt.title("1 m$^3$ of air")
plt.legend()
plt.show()
```

```python
# Initial Conditions #######################################################################################################
gamma = 1.4 # Ratio of specific heats for air (diatomic)
Patm = 101325 # Atmospheric pressure in Pa
P0 = Patm # Initial pressure in Pa
V0 = 1 # Initial volume, in m^3
Tcel = 15 # Initial temp in C

plt.figure()

T = Tcel + K # Convert to Kelvin
V = np.linspace(0.19, 1.5)
Pisotherm = (N*Kb*T)/V
Padiabat = P0*(V0/V)**gamma
T1 = 300 + K
P300C = (N*Kb*T1)/V
    


plt.plot(V, Pisotherm/1000, label="Isothermal")
plt.plot(V, Padiabat/1000, label="Adiabatic")
plt.plot(V, P300C/1000, '--', label="300 C isotherm")

# Plot Data
plt.plot(1,101.325, 'x', label=("15 C, sea level"))
plt.xlabel("Volume (m$^2$)")

plt.ylabel("Pressure (kPa)")
plt.title("1 m$^3$ of air")
plt.legend()
plt.show()
```
